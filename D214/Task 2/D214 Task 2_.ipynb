## A. Research Question

The research question for this project is as follows: "Can a ridge regression or time series model be constructed on the silver pricing dataset?" As the question suggests, the main approach to answering this research question will be a type of linear regression model called ridge regression, which will be used to attempt to predict silver prices based on macroeconomic factors. A time series model will also be used to create a prediction model, although it is outside the scope of the project requirements.

The goals of this study, made plain, are to predict silver prices using macroeconomic factors and identify the macroeconomic factors that most contribute to the fluctuation of silver prices. Predicting the price of silver is crucial for certain businesses because macroeconomic factors and historical trends in the price of silver can be used to predict the future price of silver, allowing the cost of raw materials to be minimized and the bottom line to be maximized. Furthermore, identifying which macroeconomic factors most influence the price of silver can help a business be proactive in ordering raw materials while they are still cheap.

The context selected for this question includes that the company proposing this question be answered is a solar panel manufacturing company. For such a company, silver is an important raw material. The manufacturing of solar panels relies heavily on silver powder, a material whose conductivity is highly efficient for the collection and transfer of the electrons generated when light strikes silicon (*Silver and Solar Technology*, n.d.). Because of this reliance on silver, it would be highly beneficial for a solar panel manufacturer to be able to try to predict the price of silver--or at least know which factors are predictive of the price.

For this analysis, the following null and alternative hypotheses have been established:

**Null Hypothesis (H₀):** A predictive ridge regression model with a model accuracy greater than 70% cannot be constructed on the market research dataset.

**Alternative Hypothesis (H₁):** A predictive ridge regression model with a model accuracy greater than 70% can be constructed on the market research dataset.

For the purposes of defining the "model accuracy" of the ridge regression model as it pertains to accepting or rejecting the null hypothesis, mean absolute percentage error (MAPE) will be used. This is very similar to mean absolute error, except that MAPE is expressed as a percentage by dividing the MAE by the actual value.

Measures such as mean absolute error (MAE) and root mean squared error (RMSE) are not expressed as a percentage, despite being very common ways to measure model accuracy. A percentage is easier for the layperson to understand, and one of the few cons of using MAPE, the fact that it breaks down if the actual value is 0, is accounted for by the fact that this is price data. The price of silver, unless the way of the world is flipped overnight, should never be 0 (Gilbert, 2023). While MAPE will be the primary metric for determining whether or not to accept or reject the null hypothesis, measures such as RMSE will not be disregarded. These measures will be considered alongside MAPE in order to conduct a thorough evaluation of the model's performance.

Since MAPE is a measure of error, the hypotheses above actually present "model accuracy" in an inverse manner. MAPE is best when it is as low as possible, representing predictions with smaller deviations from the actual values. Thus, in order to align with the wording of the hypotheses, a calculation of 100% minus MAPE will be used. This means that a MAPE of less than 30% is required to reject the null hypothesis. There is no universal industry standard for MAPE. It varies too much by industry for a standard to be pinpointed (Bobbitt, 2021). Thus, the threshold of 30% is chosen as a reasonable benchmark for indicating a "fair" performance of the model.

A ridge regression model was chosen for the hypotheses based on a previous study from the University of Melbourne, Australia. It found that using ridge regression is viable as a way to predict commodity prices from macroeconomic data and performs better than other regression methods because it reduces the effects of parameter estimation error by using shrinkage (Gargano & Timmermann, 2014). In addition, ridge regression does not require as many assumptions to be met as other regression models do. As an example, ridge models can handle data that is somewhat multicollinear, meaning this assumption, in particular, need not be met.

### Matching Data Frequency

This method was selected after much research, which suggested that linear interpolation was a valid method for filling missing values in specifically time-series data, which is what this dataset is, despite the main analysis for this capstone being a form of regression (GeeksforGeeks, 2024). To interpolate between the values, the "interpolate" function from pandas was used after resetting the index so that it contained one row for every day between the start and end dates of the dataframe.

Interpolation is what one might consider a "technique" used on the data, so for this, advantages and disadvantages will be given. The advantage of this technique is that it is extremely simple to implement, and it is commonly used on time series data, as recommended by the source above. It is also computationally inexpensive to use, making it well-suited to larger datasets.

The main disadvantage of this technique is that it operates on the assumption that the data changes smoothly and at a constant rate between observed data points, which may not reflect reality. In the real world, the unemployment rate may fluctuate unpredictably between actually recorded dates, meaning interpolation could oversimplify or hide short-term data variability.

### Merging Dataframes

Below, the dataframes were merged into one large dataframe called "combined_df" using the reduce function from functools to make it simpler, as previously discussed. To merge the dataframes, the merge function from pandas was used to perform a full outer join on all the dataframes.

Using a full outer join was necessary because it ensured all rows from every dataframe were retained in the final dataframe. A full outer join searches for a match on the specified column--in this case, the date column--and appends the new data from the dataframe to be merged in a new set of columns. If a date is missing in one dataframe but is present in another, NaN values are inserted in the missing columns. However, since the date range was standardized in the previous step, there should be no new NaN values introduced by this merge.

```python
dataframes = [LBMA_df, Exch_rt_df, Unemp_day_df, Inf_df, RecessUS_df, RecessCHN_df]

# Merge dataframes on the date column
combined_df = reduce(lambda left, right: pd.merge(left, right, on='Date', how='outer'), dataframes)
combined_df.sort_values
pd.set_option('display.max_rows', None)
combined_df.tail(10)
```

### Handling Missing Data Due to Measurement Gaps

Before deciding how to handle missing data, a brief exploratory analysis was required. First, the missingno package was used to plot a bar graph showing how sparse each column in the combined dataset was. A solid black bar is ideal and represents complete data, whereas a bar made up of many lines like a barcode indicates sporadic missing data. The plot revealed that the Silver_USD, Silver_GBP, Silver_Euro, US_AUD_rate, and inflation columns had sporadic missing data. Unemployment, US_Recession, and China_Recession were solid, indicating that there is no sporadic missing data. The plot also showed that the bars were different heights, which demonstrates something that was already known--the merged datasets only have data for a certain range of dates due to the availability of certain data.

Because it was required in the proposal, the dataframe sparsity was also calculated, which allows a high-level view of just how much data the dataframe is currently missing. The combined dataset had 45.36% sparsity, but some columns contributed disproportionately to this figure. The most problematic columns were Silver_Euro and inflation, which both suffered from sporadic missing values and a very limited date range.

Since differing date ranges were a key contributor to missingness, a function was written to retrieve the date range represented by each column. Then, the earliest date and the latest date shared by all columns except the inflation column were selected, and a "largest possible date range" was identified. This date range was 1/2/1968 through 4/21/2022. All rows outside of this range were dropped because there was no way to recover missing values for columns that lacked data before or after these dates. In most cases, the reason for missing data prior to a certain date in any column is that the data was simply not measured before this date. Imputing data here would be improper because there are no surrounding numbers to use to make a "guess" at what the missing data might have been. Any attempt at imputation here would constitute extrapolation, which assumes trends beyond the observed data.

The advantage of choosing truncation to handle this type of missing data is that it prevents the introduction of false trends that have no real basis in reality and real, recorded data. ### Handling Missing Data Due to Sporadic Sparsity

After handling outliers, the columns previously mentioned that have sporadic sparsity were treated for missing data using linear interpolation. Namely, these columns are: Silver_USD, US_AUD_rate, and inflation.

The advantages and disadvantages of linear interpolation were previously discussed in the context of converting monthly data to daily data, and those same advantages and disadvantages apply here as well. 

However, in order to avoid extrapolating outside of the date range for which each column was recorded, missing data only within these date ranges should be imputed using linear interpolation. As previously mentioned, imputing data outside of these ranges would be improper, because there are no surrounding datapoints to use to make a "guess" at what the missing data might have been. Any attempt at imputation here would constitute extrapolation, which assumes trends beyond the observed data. Thus, the date ranges of each column were identified first using a custom function and printed using a for loop.

With the date range for each column identified, the pandas function "interpolate" was used to perform linear interpolation for non-categorical columns.

For categorical columns, a forward-fill method was chosen, which simply copies the last observed value to the missing cell. Because the two columns on which the forward-fill method was performed deal with recording recessions as a binary flag, this method was appropriate because recessionary periods do not simply disappear overnight. If the previous day was in recession, it is more than likely the next day was too. At worst, this method may extend a recessionary period by a few days more than its actual duration. Forward-filling was performed using the pandas function "fillna."

Using forward-filling for missing data is a choice in technique. An advantage is that forward-filling does not rely on arbitrary assignments like mode-filling might, where the most common value is simply imputed where the missing datapoint is without regard to time order. For time-series data, mode-filling does not make much sense, as it can lead to a recession flipping on or off for a day in the middle of a larger period. A disadvantage is that if there is an extended period of missing values, this method assumes no change until the next known value, which could lead to a large extension of the duration of the recession or non-recession.

However, based on the missing data plot, it has been established that this should not be a problem as long as imputation is not performed outside of the date range for which recessionary periods were recorded since there were few, if any, missing datapoints in these columns.

With this step completed, the dataset is finalized and saved as a csv for evaluator perusal.

### Data Exploration and Justification
The data analysis technique that was employed below is ridge regression. Ridge regression analysis uses a type of linear regression to describe the contribution of each independent variable in predicting the dependent variable, Silver_USD. Ridge regression is particularly useful when high multicollinearity is present or when the data does not follow a normal distribution (NCSS Statistical Software, n.d.). Both of these issues are observed in this dataset.

To confirm that the dataset is not normally distributed, two techniques were used:
* Q-Q Plot: A Q-Q plot was generated to visually assess whether or not the data is normal. If the data were normally distributed, the points would closely follow the red diagonal line. However, the Q-Q plot for Silver_USD deviates significantly from this line, indicating non-normality.
* Shapiro-Wilk Test: A Shapiro-Wilk test is a formal test for normality that provides a test statistic and a p-value. A p-value greater than 0.05 indicates that the data is normal, while a p-value less than 0.05 suggests non-normality. The p-value for this test was 0.0000, which indicates not that the p-value is literally zero, but that it is incredibly small--so small that its value is less than one millionth. Regardless, since the p-value is less than 0.05, the Shapiro-Wilk test definitively shows that the data does not follow a normal distribution.

To confirm that multicollinearity is present in the dataset, the variance inflation factor (VIF) was used. Running the VIF for each column shows absolutely absurd VIF values for the three quantitative columns--as high as 73.9. For an ordinary multiple linear regression model, VIF should not exceed 10. Thus, there is high multicollinearity in this dataset.

The justification for ridge regression over other types of regression is related to its advantages. Ridge regression reduces the impact of multicollinearity because it applies a penalty term to large coefficients, making it more stable than ordinary multiple linear regression. This is known as L2 regularization. In addition, as previously stated, ridge regression does not assume normality--meaning it has two fewer assumptions than multiple linear regression. Fewer assumptions make ridge regression a more versatile technique since it can be used on more datasets than multiple linear regression.

Ridge regression does, however, have a disadvantage. L2 regularization shrinks coefficients to zero, but never to absolute zero--which would remove the explanatory variable from the model. Because of this, ridge regression does not perform feature selection like other techniques can, namely, lasso regression. Feature selection is another way of dealing with multicollinearity (Murel, 2024). Lasso regression, which performs L1 regularization, however, would reduce the coefficients of the explanatory variables with multicollinearity to 0, though, removing three of five variables from this analysis. At that point, this would hardly be an analysis.

Finally, ridge regression was chosen because a similar technique was used on other commodities in a study from the University of Melbourne, Australia. It found that using ridge regression is viable as a way to predict commodity prices from macroeconomic data and performs better than other regression methods because it reduces the effects of parameter estimation error by using shrinkage (Gargano & Timmermann, 2014).

A few more exploratory steps were taken below besides the Q-Q plot, Shapiro-Wilk test, and VIF table. This includes running a correlation matrix to check for correlation between columns--none of which seemed particularly high, with values never exceeding ±0.5. These values are on a scale of -1 to +1, with a value of 1 or -1 indicating perfect correlation. High correlation between columns can cause a poor analysis.

### Initial Ridge Regression Model

An initial ridge regression model was constructed below to test the baseline performance of this method. The data, stored in "ridge_df," was first split into independent variables (X) and the dependent variable (y.) Then, the data was split so that 80% of the data formed the training set, while the remaining 20% was allocated to the test set. This was done prior to scaling to prevent data leakage.

Once the data was split, StandardScaler was used to scale the independent variables. This scaler standardizes each variable by subtracting the mean from the value and dividing the result by the standard deviation, ensuring all variables have a mean of 0 and a standard deviation of 1. Scaling data is crucial for linear regression models, including ridge regression, because it ensures that all independent variables are on a similar scale, preventing larger-magnitude features from dominating smaller ones. Ridge regression, in particular, applies L2 regularization, which can unfairly penalize features with larger numerical ranges if they are not scaled. This justification is also StandardScaler's advantage: it should improve the performance of machine learning models that are particularly sensitive to feature scales without disturbing the original distribution of the data. A disadvantage specific to StandardScaler is that it does not perform as well when data does not follow a normal distribution (Nalcin, 2022). Despite this limitation, in this analysis, testing each scaler showed that StandardScaler outperformed MinMaxScaler, which tends to perform better on non-normal data. Thus, StandardScaler was selected for this ridge regression model.

After the data was scaled, a ridge regression model was fitted on the scaled training dataset, then the model was used to predict on the scaled test dataset. Several performance measures were then extracted based on predictions on the test dataset:

* Mean Absolute Percentage Error (MAPE)
* Mean Absolute Error (MAE)
* Mean Squared Error (MSE) 
* Root Mean Squared Error (RMSE)

MAPE is particularly relevant since it will be used later to determine whether to accept or reject the null hypothesis. Thus, for this initial model, the focus will be on this metric. 

Without any kind of adjustment to this very typical ridge regression model with alpha equal to 1, the performance did not impress. The initial model's MAPE was 31.08%, slightly exceeding the threshold required to reject the null hypothesis. To improve performance, adjustments are necessary, and fortunately, there are several techniques an analyst can employ. One such technique is applied below: adding polynomial features.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
```

```python
# Split data into train and test sets before scaling to prevent data leakage
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
                                    X, y, test_size=0.2, random_state=7)
```

```python
# Ridge Regression model without polynomial features
# Hm. not the best model.
scaler = StandardScaler()
X_tr_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

ridge = Ridge(alpha=1)
ridge.fit(X_tr_scaled, y_train)

y_pred = ridge.predict(X_test_scaled)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = ridge.score(X_test_scaled, y_test)


print(f"Mean Absolute Error (MAE): ${mae:.4f} per troy oz")
print(f"Mean Squared Error (MSE): ${mse:.4f} per troy oz²")
print(f"Root Mean Squared Error (RMSE): ${rmse:.4f} per troy oz")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.4%}")
print(f"R-Squared: {r2:.4f}")
```

### Model Adjustments using Polynomial Features

In order to determine which adjustment method might be most useful for improving the performance of the ridge regression model, seaborn's pairplot was used to visualize how the quantitative explanatory variables related to the dependent variable. Ideally, these plots should reveal a linear relationship, as ridge regression relies on the existence of a linear relationship to be able to make accurate predictions. However, the plots showed what is more aptly called chaos, especially in the instance of the unemployment rate versus the silver prices, which appeared highly scattered.

Despite these oddities, some variables did show structure--notably, the USD to AUD exchange rate and the inflation rate, both of which followed a curvilinear trend, appearing somewhat parabolic. Because of this, it seemed justified to attempt using polynomial features from sklearn to try and capture these more complex relationships in the data.

This method generates new features by raising the original variables to different powers (inflation², inflation³, etc.,) allowing the model to capture non-linear trends. In this case, polynomial features up to the fourth degree were added to the ridge regression model. Polynomial features also, by default, adds in terms that represent the interaction of two different features such as unemployment x inflation. This allows the model to consider how the effect of one feature might depend on the value of another.

An advantage of polynomial regression is related, again, to the justification for its use in this analysis. It allows the analyst to fit a wide range of curvilinear relationships beyond simple linear ones, allowing the model to capture more complex trends than would otherwise be possible. A disadvantage is that it can be highly sensitive to outliers, which seriously affects the accuracy of the model by distorting higher-degree terms (GeeksforGeeks, 2024a). In addition, adding polynomial terms grows the model's complexity exponentially, making it harder to interpret and more likely to overfit. In the case below, the number of explanatory variables exploded from just 5 to 125 after polynomial features were implemented. The interaction terms are especially hard to interpret.

Below, polynomial features up to degree four were applied to the X_train and X_test datasets. As mentioned previously, the big disadvantage with ARIMA/SARIMAX over ridge regression is that it does not tell the analyst why the price has changed and provides no market conditions to "look out for" that might warn of a rise in the price of silver.

#### ARIMA Exploration: Stationarity

Before an ARIMA can be run, some data exploration must be conducted to determine a few factors that influence what type of time series model will be used. These factors include stationarity, seasonality, trends, and residuals.

Stationarity is a key assumption for the ARIMA model, meaning the data's mean and variance should not change depending on time (Singh, 2023). Thus, the time series data will be visualized in a line graph as a preliminary step in determining if the data is stationary. Data that is not stationary can be made stationary by differencing the dataset a number of times until it shows non-stationarity. In this case, it is obvious the data is non-stationary due to the slowly creeping yet sustained upward trend in the line graph. However, to verify this visual analysis, an Augmented Dickey-Fuller (ADF) test for stationarity will be conducted. Performing an ADF test will formalize the assumption from the line graph that the data is not stationary.

When performing an ADF test, the p-value signifies whether the data is stationary or not. If $p < 0.05$, the null hypothesis (H₀) specified below should be rejected, meaning the time series is stationary. If $p > 0.05$, the null hypothesis should fail to be rejected, meaning the time series is non-stationary.

*H₀ = The time series data is non-stationary.*

*H₁ = The time series data is stationary.*

Below, an ADF test on the original time series dataset was performed first, which returned a p-value of 0.12, which is well above the threshold to fail to reject the null hypothesis. Thus, the data is non-stationary. At this point, it was necessary to difference the data until it was stationary. Doing so indicates what the ARIMA $d$ term should be.

An advantage to running an ADF test instead of just relying on visual analysis is that it is far more objective, with no room for uncertainty in the result. It is simple to interpret. A disadvantage is that since it is a statistical test, it may struggle with very large, noisy, or complex datasets (Jones, 2025).

After the data was differenced once, the differenced data was graphed and an ADF was run. This time, the graph does not show any upward or downward trends--it is noisy but relatively flat. The ADF test concurs, with a result of 0.0. This does not mean the p-value is *literally* 0.0, only that it is very small. Regardless, the results of the ADF test confirm that the data, after being differenced once, is now stationary.

#### ARIMA Model: Construction

The data was split so that 80% of the data formed the training set, while the remaining 20% was allocated to the test set. In the call to "train_test_split," it is essential to set shuffle equal to "False" so that the data isn't randomly reordered for the split, since time series analysis depends on time-ordered data. Different from how the data was split for ridge analysis, the result of "train_test_split" here is only two datasets: one training set, and one testing set. For ridge analysis, there were four, two for training and two for testing split between the X variables and y variable. This again highlights the difference between time series models and regression: regression explains y using the X variables whereas time series uses patterns in the one column it's given to predict values of that column out into the future.

This split corresponds to a roughly 11 year long test dataset, while the train dataset is roughly 43 years long (these numbers are rounded by quite a lot for simplicity's sake.) However, ARIMA is not ideal for long term forecasting, especially not beyond 6 months (Hayes, 2024). Therefore, any subsequent forecast performed using the test data will be performed only on 6 month intervals of the test data to ensure the model is not overextended.

The "auto_arima" function from the pmdarima package was then run to determine the optimal p, d, and q values--in other words, the optimal order for the ARIMA model. This function can perform either a grid search or a stepwise search. A grid search analyzes all possible combinations of p, d, and q values up until a stopping point set by the user. While this approach is exhaustive, it is also very time-consuming. Instead of this, a stepwise search was performed, which finds the best model by adding or removing AR, I, or MA terms one at a time until the AIC value stops improving or flattens out. In this process, it minimizes AIC. The lower the AIC, the better the model.

Stepwise search is significantly faster than grid search unless the search space is very small, which is its advantage. Given that the PACF and ACF graphs suggested an ARIMA model with a minimum order of (3,1,3)--and quite possibly higher $p$ and $q$ values--the search space is already quite large. A grid search evaluating all models up to (3,1,3) would need to test 16 different models while extending the search to (5,1,5) increases this number to 36. These numbers do not account for variations in seasonal components or whether an intercept is included, which would further expand the search space. Because stepwise search strategically selects models to test based on performance, it avoids being time-consuming and computationally intensive.

However, the major disadvantage to stepwise search is also that it's not exhaustive--the very thing that makes it efficient. It doesn't test every possible model and may very well overlook a better-performing model that a grid search would have found. Therefore, when choosing a process, it is important to understand which is more important: efficiency or performance. For this project, efficiency was chosen over performance, given that the stepwise search still took 10 minutes to complete.

The results of auto_arima suggest that a (5,1,5) model with no intercept or seasonality performed the best. For comparison, the AIC of both the suggested model and the model indicated by the PACF and ACF graphs (indicated model) are printed. The model suggested by auto_arima achieved an AIC of 6789.87 while the indicated model achieved an AIC of only 6898.26. Because auto_arima's model has an AIC that is significantly lower than the indicated model, the auto_arima model's order was chosen for the next step: constructing the model in earnest.

Using statsmodels' "ARIMA" function, an ARIMA model was fit to the training dataset with the order (5,1,5) and the model summary was printed.

```python
# Time to split--the original arima_df. ARIMA will take care of
# differencing itself. But we now know what value to expect for d...1.

train, test = train_test_split(arima_df, test_size = 0.2, shuffle = False, random_state=37)
```

## E. Data Summary and Implications

As a refresher, the research question and hypotheses are printed below.

**Research Question:** Can a ridge regression or time series model be constructed on the silver pricing dataset?

**Null Hypothesis (H₀):** A predictive ridge regression model with a model accuracy greater than 70% cannot be constructed on the market research dataset.

**Alternative Hypothesis (H₁):** A predictive ridge regression model with a model accuracy greater than 70% can be constructed on the market research dataset.

As it pertains only to the research question and its hypotheses, MAPE is the sole metric by which performance will be measured. As mentioned previously, since MAPE is a measure of error, the hypotheses above actually present "model accuracy" in an inverse manner. Thus, in order to align with the wording of the hypotheses, a calculation of 100% minus MAPE will be used. This means that a MAPE of less than 30% is required to reject the null hypothesis.

The final ridge regression model with polynomial features achieved a MAPE of 21.77%, which, when converted to how the research question is worded, corresponds to 78.23% accuracy. The final time series model achieved a MAPE of 10.05%, which is an 89.95% accuracy. Using only this metric, the null hypothesis can be rejected in favor of the alternate hypothesis, as the ridge model easily clears the threshold set by the hypotheses. However, there is still more to discuss in regards to performance--and how a "good" MAPE can be misleading.

The table below provides the final metrics for each model.
