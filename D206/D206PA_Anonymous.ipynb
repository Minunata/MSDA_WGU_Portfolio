### A: Research Question

For the medical dataset, my research question is "What are the major contributing factors that lead to patient readmission?" It is generally advisable that a hospital's readmission rate be low for several reasons. A low readmission rate can improve patient outcomes and patient satisfaction while a high readmission rate puts increased strain on physicians, wastes time and resources, and lowers patient satisfaction scores (Regis College, 2022). This research question can help a hospital identify the leading causes of patient readmission so that a more proactive approach to reducing patient readmission can be taken in the future. While patients may have many factors that increase readmission risk and those factors may be uncontrollable, greater care can be taken with such patients to improve the factors the hospital can control.

## Part II: Data Cleaning Plan (Detection)
### C1: Detection Plan
In order to identify duplicates, first I will pull out the columns I wish to assess the uniqueness of and store them in a new dataframe so I can use a for loop to quickly iterate through them and get the information I wish to know in fewer lines of code. I will use value_counts().count() to get a count of the rows in these columns, which I can compare to an expected number. In addition to that, I will double-check for duplicates using .duplicated().value_counts() which will return the number of true and false rows. If there are no duplicates, this should return no trues.

For missing values, I will use missingno's .matrix to create a missing data matrix and matplotlib.pyplot's .title() and .show() to customize the matrix. It is worth noting that .info() will also help provide information on missing values by showing the number of not null values in each column, but it will not be the primary source of information on missing data, rather, it is more of a sanity check. I will gather more information about the missing data by creating a new dataframe called df_nulls and use .isnull() to store all of the original dataframe's nulls in it. Once stored, I can then use .mean() on this dataframe, which will tell me what percentage of each column is null. In order to give even more detail, I will also use .sum() on df_nulls to find the number of nulls in each column.

On quantitative variables only, I will apply the following approach to identify statistical outliers. First, I will define a new function called find_outliers_IQR() that will help me better understand the boxplots I wish to create later on. Sometimes the boxplots are hard to read if there are many, many outliers, so some numerical data is needed. Find_outliers_IQR() mimics how seaborn's .boxplot() identifies outliers. It calculates the first and third quartiles, then the interquartile range. After the IQR is calculated, it calculates two equations: Q1 - (1.5 * IQR) and Q3 + (1.5 * IQR). These are stored in outliers, which is then returned by the function. With the new function defined, these are the steps I followed to get information about the presence, number, and values of outliers. First, I will use .describe() to do a quick visual check for clear outliers, indicated by the min and max and other statistics returned. Second, I will use seaborn's .boxplot() to create a plot that shows outliers as dots on either side of the "whiskers" of the graph. This will allow me to identify if outliers are present at all. If outliers are present, I will continue by feeding my newly defined find_outliers_IQR() function the column I am interested in and store the output in an appropriately named new dataframe such as outliers_pop for Population outliers. Once stored, I can query this dataframe using len() to return the number of outliers in the column (to print it out nicely, I also used str() to convert the column length returned by len() to a string.) Similarly, I can query for the min and max outliers using .min() and .max() respectively (these were also converted to strings for print() purposes.) In order to find the exact values of each and every outlier, I will simply print the dataframe containing the outliers. In some cases, I will use .sort_values() to sort the dataframe before printing it if there are outlier ranges on both the lower and upper ends of the boxplot. This allows me to visually see where the ranges begin and end so I can report them.

Lastly, I will check for incorrect datatypes, categorical variables that need re-expression or re-encoding, and categorical variables that have options that would be better if combined. To check for incorrect datatypes, I will use .info() on the dataframe containing all my data to get a summary of the columns and their datatypes. I can quickly scan this for obvious errors. In general, to check for categorical variables that need re-encoding and categorical variables that have superfluous options, I used .value_counts(). This returns the total number of rows for each option, which is a great summary to scan for problems. The option name is provided in this summary, making it easy to see the categorical variables that need to be re-encoded.

```python
# define function for outlier detection throughout the following code. [In-Text Citation: (Kleppen, 2023).]
def find_outliers_IQR(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3 - q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers
```

### D4: Treatment Code
Below you will find the code used to treat anomalies such as improper datatypes, missing data, duplicate rows, and outliers.
```python
# Copy dataframe to store changes, just in case I need the original.
df_clean = df.copy()
# I will deal with items 1-29 in section D1 above first.
# Convert Zip to string from integer.
df_clean['Zip'] = df_clean['Zip'].astype('str')
# Add leading zeros using zfill()
df_clean['Zip'] = df_clean['Zip'].str.zfill(5)
# Identify columns that can be converted all at once to category datatype using for loop.
category_cols = df_clean[['Area', 'Marital', 'Initial_admin', 'Complication_risk', 'Services', 'Education', 'Employment', 'ReAdmis',
                         'Soft_drink', 'HighBlood', 'Stroke', 'Arthritis', 'Diabetes', 'Hyperlipidemia', 'BackPain',
                         'Allergic_rhinitis', 'Reflux_esophagitis', 'Asthma']]
# Will do Item# columns later since they require an order. Timezone and Gender need a dict written, will do that later too.
# Anxiety and Overweight need re-encoding, will do those separately as well.
for col in category_cols:
    df_clean[col] = df_clean[col].astype('category')
# Write dictionary for option reduction for Timezone column
mapping_timezone = {'America/Puerto_Rico' : 'Atlantic',
          'America/New_York' : 'Eastern',
          'America/Detroit' : 'Eastern',
          'America/Indiana/Indianapolis' : 'Eastern',
          'America/Indiana/Vevay' : 'Eastern',
          'America/Indiana/Vincennes' : 'Eastern',
          'America/Kentucky/Louisville' : 'Eastern',
          'America/Toronto' : 'Eastern',
          'America/Indiana/Marengo' : 'Eastern',
          'America/Indiana/Winamac' : 'Eastern',
          'America/Chicago' : 'Central',
          'America/Menominee' : 'Central',
          'America/Indiana/Knox' : 'Central',
          'America/Indiana/Tell_City' : 'Central',
          'America/North_Dakota/Beulah' : 'Central',
          'America/North_Dakota/New_Salem' : 'Central',
          'America/Denver' : 'Mountain',
          'America/Boise' : 'Mountain',
          'America/Phoenix' : 'Mountain',
          'America/Los_Angeles' : 'Pacific',
          'America/Nome' : 'Alaskan',
          'America/Anchorage' : 'Alaskan',
          'America/Sitka' : 'Alaskan',
          'America/Yakutat' : 'Alaskan',
          'America/Adak' : 'Hawaiian',
          'Pacific/Honolulu' : 'Hawaiian'
          }
# Use dictionary to convert timezone options.
df_clean.Timezone.replace(mapping_timezone, inplace=True)
# Convert timezone to category datatype.
df_clean['Timezone'] = df_clean['Timezone'].astype('category')
# Write dictionary option reduction for Gender
mapping_gender = {'Prefer not to answer' : 'Non-binary'}
# Use dictionary to convert Gender options.
df_clean.Gender.replace(mapping_gender, inplace = True)
# Convert gender to category datatype.
df_clean['Gender'] = df_clean['Gender'].astype('category')
# Write dictionary to re-encode 1 and 0 to yes and no for Overweight and Anxiety columns.
mapping_overwt_anx = {1.0 : 'Yes',
          0.0 : 'No',
          }
# Use dictionary to convert Overweight options.
df_clean.Overweight.replace(mapping_overwt_anx, inplace = True)
# Convert Overweight to category datatype.
df_clean['Overweight'] = df_clean['Overweight'].astype('category')
# Use dictionary to convert Anxiety options.
df_clean.Anxiety.replace(mapping_overwt_anx, inplace = True)
# Convert Overweight to category datatype.
df_clean['Anxiety'] = df_clean['Anxiety'].astype('category')
```

## Part IV: PCA
### E1: Principal Components
For the purposes of PCA, only quantitative variables can be used. These are as follows: Lat, Lng, Population, Children, Age, Income, VitD_level, Doc_visits, Full_meals_eaten, VitD_supp, Initial_days, TotalCharge, and Additional_charges. I will leave out CaseOrder since it is an index. As shown in the loadings matrix below, there are 13 total principal components, though not all may be retained.

```python
from sklearn.decomposition import PCA
import numpy as np
# Isolate quantitative columns from the other columns in the dataset.
df_pca = df_clean[['Lat', 'Lng', 'Population', 'Children', 'Age', 'Income', 'VitD_levels', 'Doc_visits', 
                   'Full_meals_eaten', 'VitD_supp', 'Initial_days', 'TotalCharge', 'Additional_charges']]
# Normalize columns by subtracting the mean and dividing by standard deviation.
df_pca_norm = (df_pca - df_pca.mean())/df_pca.std()
# Set number of components to be returned equal to number of variables in numeric-only dataset.
pca = PCA(n_components = df_pca.shape[1])
# Fit normalized dataset to PCA.
pca.fit(df_pca_norm)
# Create dataframe and name columns to better be able to read the PCA.
pca_print = pd.DataFrame(pca.transform(df_pca_norm), columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8',
                                                               'PC9', 'PC10', 'PC11', 'PC12', 'PC13'])
# Create dataframe to house PCA loadings and print them.
loadings = pd.DataFrame(pca.components_.T, columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 
                                                      'PC10', 'PC11', 'PC12', 'PC13'], index = df_pca_norm.columns)
loadings
```

### E2: Retained Principal Components
I will be using the Kaiser method to decide which principal components to keep. This rule states that if the eigenvalue for a principal component is equal to or greater than one, it should be kept. To identify PCs with an eigenvalue greater than or equal to one, I will visually examine a scree plot. As shown below, the plot is a bit hard to read at PC6, so to determine whether or not to keep PC6, I will print the eigenvalues instead.

```python
# Create covariance matrix
covar_matrix = np.dot(df_pca_norm.T, df_pca_norm)/df_pca.shape[0]
# Generate eigenvalues
eigenvalues = [np.dot(eigenvector.T, np.dot(covar_matrix, eigenvector)) for eigenvector in pca.components_]
# Generate scree plot of eigenvalues, with a line at y = 1 to help figure out which PCs to keep.
plt.plot(eigenvalues)
plt.xlabel('# of Components')
plt.ylabel('Eigenvalues')
plt.axhline(y = 1, color = 'red')
plt.show()
```

The uncertain values of PC6 and PC7 are clearly specified above, and PC6 is above 1 while PC7 is not. Thus, using the Kaiser rule, PCs 1 through 6 will be kept. PCs 7 through 13 have eigenvalues below 1, so there is no need to keep them. Eigenvalues represent the amount of variance explained by each PC, and any PC with an eigenvalue below 1 does not explain enough variance to be kept.

While there were not many quantitative variables to do PCA on in this dataset, it did allow us to halve the number of variables used in PCA. Reducing dimensionality is beneficial to organizations because high dimensionality is often detrimental to the performance of algorithms used to analyze the data, causing  them to run slowly and require significantly more resources (Biga Bid Ltd, 2023). Sacrificing a bit of accuracy is worth the time saved, especially if the amount of accuracy sacrificed is small and the dimensionality is much reduced. Another benefit of PCA is that it can help to remove unnecessary noise in the data that could lead to poor analysis and its ability to provide an organization with uncorrelated data features (Biga Bid Ltd, 2023).
